```python
# ============================================================
# Hybrid Deep Learning + q-Rung Orthopair Fuzzy Diagnosis
# Stroke Prediction Dataset - Google Colab Ready
# ============================================================

# ---------------------------
# 1. Mount Google Drive
# ---------------------------
from google.colab import drive
drive.mount('/content/drive')

# ---------------------------
# 2. Imports
# ---------------------------
import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    RocCurveDisplay
)
from sklearn.utils.class_weight import compute_class_weight

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.utils import plot_model

# ---------------------------
# 3. Set random seeds
# ---------------------------
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ---------------------------
# 4. File path from Google Drive
# ---------------------------
# Change this path to your actual file location in Drive
FILE_PATH = "/content/drive/MyDrive/Dataset /Fuzzy/healthcare-dataset-stroke-data.csv"

if not os.path.exists(FILE_PATH):
    raise FileNotFoundError(f"CSV file not found at: {FILE_PATH}")

# ---------------------------
# 5. Load dataset
# ---------------------------
df = pd.read_csv(FILE_PATH)

print("Dataset loaded successfully.")
print("Shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())

print("\nColumn names:")
print(df.columns.tolist())

# ---------------------------
# 6. Basic cleaning
# ---------------------------
# Remove ID column if present
if 'id' in df.columns:
    df = df.drop(columns=['id'])

# Standardize a common typo if found
if 'work_type' in df.columns:
    df['work_type'] = df['work_type'].replace({'Govt_jov': 'Govt_job'})

print("\nMissing values:")
print(df.isnull().sum())

# ---------------------------
# 7. Separate features and target
# ---------------------------
TARGET_COL = 'stroke'

if TARGET_COL not in df.columns:
    raise ValueError(f"Target column '{TARGET_COL}' not found in dataset.")

X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL].copy()

print("\nTarget distribution:")
print(y.value_counts())

# ---------------------------
# 8. Identify numeric and categorical columns
# ---------------------------
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

print("\nNumeric features:", numeric_features)
print("Categorical features:", categorical_features)

# ---------------------------
# 9. Preprocessing pipeline
# ---------------------------
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', tf.keras.layers.StringLookup)  # placeholder comment only
])

# We will use pandas.get_dummies for simplicity after imputation
# First impute missing values manually for a cleaner Colab workflow

for col in numeric_features:
    df[col] = df[col].fillna(df[col].median())

for col in categorical_features:
    df[col] = df[col].fillna(df[col].mode()[0])

# One-hot encode categorical variables
df_encoded = pd.get_dummies(df, columns=categorical_features, drop_first=True)

X = df_encoded.drop(columns=[TARGET_COL])
y = df_encoded[TARGET_COL].astype(int)

# Scale numerical features after encoding
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nEncoded feature shape:", X_scaled.shape)

# ---------------------------
# 10. Train-test split
# ---------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=SEED,
    stratify=y
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train,
    y_train,
    test_size=0.2,
    random_state=SEED,
    stratify=y_train
)

print("\nTrain shape:", X_train.shape)
print("Validation shape:", X_val.shape)
print("Test shape:", X_test.shape)

# ---------------------------
# 11. Handle class imbalance
# ---------------------------
classes = np.unique(y_train)
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=y_train
)
class_weight_dict = {cls: weight for cls, weight in zip(classes, class_weights)}

print("\nClass weights:", class_weight_dict)

# ---------------------------
# 12. Build Hybrid Deep Learning model
# ---------------------------
input_dim = X_train.shape[1]

inputs = layers.Input(shape=(input_dim,), name="Input_Features")

x = layers.Dense(128, activation='relu')(inputs)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)

x = layers.Dense(64, activation='relu')(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.25)(x)

# Deep feature layer
deep_features = layers.Dense(32, activation='relu', name="Deep_Feature_Layer")(x)

x = layers.Dense(16, activation='relu')(deep_features)
outputs = layers.Dense(1, activation='sigmoid', name="Stroke_Probability")(x)

model = models.Model(inputs=inputs, outputs=outputs, name="Hybrid_DL_qROF_Model")

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
)

model.summary()

# ---------------------------
# 13. Train model
# ---------------------------
early_stop = callbacks.EarlyStopping(
    monitor='val_auc',
    mode='max',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    class_weight=class_weight_dict,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

# ---------------------------
# 14. Plot training history
# ---------------------------
plt.figure(figsize=(8, 5))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training vs Validation Accuracy')
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history.history['auc'], label='Train AUC')
plt.plot(history.history['val_auc'], label='Val AUC')
plt.xlabel('Epoch')
plt.ylabel('AUC')
plt.title('Training vs Validation AUC')
plt.legend()
plt.show()

# ---------------------------
# 15. Predict probabilities
# ---------------------------
y_prob = model.predict(X_test).ravel()
y_pred = (y_prob >= 0.5).astype(int)

# ---------------------------
# 16. q-Rung Orthopair Fuzzy conversion
# ---------------------------
# For each prediction probability p:
# membership mu = p
# non-membership nu = (1 - p^q)^(1/q) adjusted to keep mu^q + nu^q <= 1
# hesitation pi = (1 - mu^q - nu^q)^(1/q)

def qrof_transform(probabilities, q=3):
    """
    Convert scalar probabilities into q-Rung Orthopair Fuzzy values.
    We use:
        mu = p
        nu = (1 - p) * alpha
    Then enforce mu^q + nu^q <= 1
    """
    mu_list, nu_list, pi_list = [], [], []

    for p in probabilities:
        mu = float(np.clip(p, 0.0, 1.0))

        # initial non-membership from inverse confidence
        nu = float(np.clip(1.0 - p, 0.0, 1.0))

        # enforce q-ROF constraint
        if (mu ** q + nu ** q) > 1.0:
            # rescale nu safely
            max_nu = (max(0.0, 1.0 - mu ** q)) ** (1.0 / q)
            nu = min(nu, max_nu)

        pi = max(0.0, 1.0 - mu ** q - nu ** q) ** (1.0 / q)

        mu_list.append(mu)
        nu_list.append(nu)
        pi_list.append(pi)

    return np.array(mu_list), np.array(nu_list), np.array(pi_list)

# You can change q = 1, 2, 3, 4...
Q_RUNG = 3
mu, nu, pi = qrof_transform(y_prob, q=Q_RUNG)

# ---------------------------
# 17. Fuzzy score function
# ---------------------------
# Score = mu^q - nu^q
def qrof_score(mu, nu, q=3):
    return (mu ** q) - (nu ** q)

fuzzy_scores = qrof_score(mu, nu, q=Q_RUNG)

# Threshold score > 0 means more support for stroke
y_pred_fuzzy = (fuzzy_scores > 0).astype(int)

# ---------------------------
# 18. Evaluation: Deep Learning output
# ---------------------------
print("\n===== Standard Deep Learning Prediction Results =====")
print("Accuracy :", round(accuracy_score(y_test, y_pred), 4))
print("Precision:", round(precision_score(y_test, y_pred, zero_division=0), 4))
print("Recall   :", round(recall_score(y_test, y_pred, zero_division=0), 4))
print("F1-score :", round(f1_score(y_test, y_pred, zero_division=0), 4))
print("ROC-AUC  :", round(roc_auc_score(y_test, y_prob), 4))

print("\nClassification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# ---------------------------
# 19. Evaluation: q-ROF fuzzy score output
# ---------------------------
print("\n===== q-Rung Orthopair Fuzzy Diagnosis Results =====")
print("Accuracy :", round(accuracy_score(y_test, y_pred_fuzzy), 4))
print("Precision:", round(precision_score(y_test, y_pred_fuzzy, zero_division=0), 4))
print("Recall   :", round(recall_score(y_test, y_pred_fuzzy, zero_division=0), 4))
print("F1-score :", round(f1_score(y_test, y_pred_fuzzy, zero_division=0), 4))

print("\nClassification Report (Fuzzy Score Based):")
print(classification_report(y_test, y_pred_fuzzy, zero_division=0))

print("Confusion Matrix (Fuzzy Score Based):")
print(confusion_matrix(y_test, y_pred_fuzzy))

# ---------------------------
# 20. ROC Curve
# ---------------------------
RocCurveDisplay.from_predictions(y_test, y_prob)
plt.title("ROC Curve - Hybrid Deep Learning Model")
plt.show()

# ---------------------------
# 21. Save results to CSV
# ---------------------------
results_df = pd.DataFrame({
    'Actual': y_test.values,
    'DL_Probability': y_prob,
    'DL_Prediction': y_pred,
    f'qROF_mu_q{Q_RUNG}': mu,
    f'qROF_nu_q{Q_RUNG}': nu,
    f'qROF_pi_q{Q_RUNG}': pi,
    f'qROF_score_q{Q_RUNG}': fuzzy_scores,
    'Fuzzy_Prediction': y_pred_fuzzy
})

SAVE_PATH = "/content/drive/MyDrive/Dataset /Fuzzy/stroke_hybrid_qrof_results.csv"
results_df.to_csv(SAVE_PATH, index=False)

print(f"\nResults saved to: {SAVE_PATH}")
display(results_df.head(10))

# ---------------------------
# 22. Save trained model
# ---------------------------
MODEL_PATH = "/content/drive/MyDrive/Dataset /Fuzzy/hybrid_qrof_stroke_model.h5"
model.save(MODEL_PATH)
print(f"Model saved to: {MODEL_PATH}")
```

```python
# ============================================================
# Hybrid Deep Learning + q-Rung Orthopair Fuzzy Diagnosis
# Stroke Prediction Dataset - Improved Colab Version
# Includes:
# 1) Google Drive loading
# 2) SMOTE
# 3) Threshold tuning
# 4) Feature-level fuzzy mapping
# 5) Optional focal loss
# ============================================================

# ---------------------------
# 1. Mount Google Drive
# ---------------------------
from google.colab import drive
drive.mount('/content/drive')

# ---------------------------
# 2. Install required package
# ---------------------------
!pip -q install imbalanced-learn

# ---------------------------
# 3. Imports
# ---------------------------
import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report
)

from imblearn.over_sampling import SMOTE

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, backend as K

# ---------------------------
# 4. Reproducibility
# ---------------------------
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ---------------------------
# 5. File path
# ---------------------------
# Change if needed
FILE_PATH = "/content/drive/MyDrive/Dataset /Fuzzy/healthcare-dataset-stroke-data.csv"

if not os.path.exists(FILE_PATH):
    raise FileNotFoundError(f"File not found: {FILE_PATH}")

# ---------------------------
# 6. Load dataset
# ---------------------------
df = pd.read_csv(FILE_PATH)
print("Loaded shape:", df.shape)
display(df.head())

# ---------------------------
# 7. Basic cleaning
# ---------------------------
if 'id' in df.columns:
    df.drop(columns=['id'], inplace=True)

if 'work_type' in df.columns:
    df['work_type'] = df['work_type'].replace({'Govt_jov': 'Govt_job'})

# Fill missing BMI
if 'bmi' in df.columns:
    df['bmi'] = df['bmi'].fillna(df['bmi'].median())

categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing values after cleaning:")
print(df.isnull().sum())

# ---------------------------
# 8. Keep a copy before encoding
# ---------------------------
df_original = df.copy()

# ---------------------------
# 9. One-hot encode for deep learning branch
# ---------------------------
df_encoded = pd.get_dummies(df, drop_first=True)

TARGET = 'stroke'
X = df_encoded.drop(columns=[TARGET])
y = df_encoded[TARGET].astype(int)

print("\nEncoded shape:", X.shape)
print("\nTarget distribution:")
print(y.value_counts())

# ---------------------------
# 10. Train/Val/Test split
# ---------------------------
X_train_full, X_test, y_train_full, y_test, idx_train_full, idx_test = train_test_split(
    X, y, df_original.index,
    test_size=0.20,
    stratify=y,
    random_state=SEED
)

X_train, X_val, y_train, y_val, idx_train, idx_val = train_test_split(
    X_train_full, y_train_full, idx_train_full,
    test_size=0.20,
    stratify=y_train_full,
    random_state=SEED
)

# Original-row copies for fuzzy feature mapping
df_train_orig = df_original.loc[idx_train].copy()
df_val_orig   = df_original.loc[idx_val].copy()
df_test_orig  = df_original.loc[idx_test].copy()

# ---------------------------
# 11. Scale features for deep learning branch
# ---------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled   = scaler.transform(X_val)
X_test_scaled  = scaler.transform(X_test)

# ---------------------------
# 12. SMOTE on training data only
# ---------------------------
smote = SMOTE(random_state=SEED)
X_train_sm, y_train_sm = smote.fit_resample(X_train_scaled, y_train)

print("\nBefore SMOTE:", y_train.value_counts().to_dict())
print("After SMOTE :", pd.Series(y_train_sm).value_counts().to_dict())

# ============================================================
# FEATURE-LEVEL FUZZY MAPPING
# ============================================================

# We create fuzzy risk features from important clinical variables:
# - age
# - avg_glucose_level
# - bmi
# - hypertension
# - heart_disease
# - smoking_status
#
# This is better than fuzzy mapping from final probability alone.

# ---------------------------
# 13. Membership functions
# ---------------------------
def triangular(x, a, b, c):
    x = np.asarray(x, dtype=float)
    y = np.zeros_like(x, dtype=float)

    left = (a < x) & (x <= b)
    right = (b < x) & (x < c)

    if b != a:
        y[left] = (x[left] - a) / (b - a)
    if c != b:
        y[right] = (c - x[right]) / (c - b)

    y[x == b] = 1.0
    return np.clip(y, 0, 1)

def trapezoidal(x, a, b, c, d):
    x = np.asarray(x, dtype=float)
    y = np.zeros_like(x, dtype=float)

    rise = (a < x) & (x < b)
    top = (b <= x) & (x <= c)
    fall = (c < x) & (x < d)

    if b != a:
        y[rise] = (x[rise] - a) / (b - a)
    y[top] = 1.0
    if d != c:
        y[fall] = (d - x[fall]) / (d - c)

    return np.clip(y, 0, 1)

# ---------------------------
# 14. Build fuzzy features from original data
# ---------------------------
def build_fuzzy_features(df_in):
    out = pd.DataFrame(index=df_in.index)

    # Age fuzzy memberships
    out["age_young"]  = triangular(df_in["age"], 0, 20, 40)
    out["age_middle"] = triangular(df_in["age"], 30, 50, 70)
    out["age_old"]    = trapezoidal(df_in["age"], 55, 65, 100, 110)

    # Glucose fuzzy memberships
    out["glucose_normal"] = triangular(df_in["avg_glucose_level"], 60, 90, 125)
    out["glucose_high"]   = trapezoidal(df_in["avg_glucose_level"], 110, 140, 300, 350)

    # BMI fuzzy memberships
    out["bmi_normal"] = triangular(df_in["bmi"], 18, 22, 25)
    out["bmi_high"]   = trapezoidal(df_in["bmi"], 24, 28, 60, 70)

    # Binary clinical risk factors
    out["hypertension_risk"] = df_in["hypertension"].astype(float)
    out["heart_disease_risk"] = df_in["heart_disease"].astype(float)

    # Smoking risk mapping
    smoking_map = {
        "never smoked": 0.0,
        "Unknown": 0.2,
        "formerly smoked": 0.6,
        "smokes": 1.0
    }
    out["smoking_risk"] = df_in["smoking_status"].map(smoking_map).fillna(0.2)

    # Married proxy risk
    out["married_risk"] = df_in["ever_married"].map({"No": 0.0, "Yes": 1.0}).fillna(0.0)

    # Work-type risk mapping
    work_map = {
        "children": 0.0,
        "Never_worked": 0.1,
        "Govt_job": 0.4,
        "Private": 0.5,
        "Self-employed": 0.6
    }
    out["work_risk"] = df_in["work_type"].map(work_map).fillna(0.3)

    # Residence risk, mild effect
    out["urban_risk"] = df_in["Residence_type"].map({"Rural": 0.4, "Urban": 0.6}).fillna(0.5)

    return out

fuzzy_train = build_fuzzy_features(df_train_orig)
fuzzy_val   = build_fuzzy_features(df_val_orig)
fuzzy_test  = build_fuzzy_features(df_test_orig)

print("\nFuzzy feature sample:")
display(fuzzy_train.head())

# ---------------------------
# 15. Build q-ROF feature aggregation from fuzzy features
# ---------------------------
def qrof_feature_mapping(fuzzy_df, q=3):
    # Weighted fuzzy membership risk
    # Higher age_old, glucose_high, bmi_high, hypertension, heart_disease, smoking
    mu = (
        0.18 * fuzzy_df["age_old"] +
        0.15 * fuzzy_df["glucose_high"] +
        0.10 * fuzzy_df["bmi_high"] +
        0.15 * fuzzy_df["hypertension_risk"] +
        0.15 * fuzzy_df["heart_disease_risk"] +
        0.10 * fuzzy_df["smoking_risk"] +
        0.07 * fuzzy_df["married_risk"] +
        0.05 * fuzzy_df["work_risk"] +
        0.05 * fuzzy_df["urban_risk"]
    )

    mu = np.clip(mu.values.astype(float), 0, 1)

    # Protective / non-risk side
    nu = (
        0.25 * fuzzy_df["age_young"] +
        0.20 * fuzzy_df["glucose_normal"] +
        0.20 * fuzzy_df["bmi_normal"] +
        0.20 * (1 - fuzzy_df["hypertension_risk"]) +
        0.15 * (1 - fuzzy_df["heart_disease_risk"])
    )
    nu = np.clip(nu.values.astype(float), 0, 1)

    # Enforce q-rung orthopair constraint
    nu_adj = []
    pi_adj = []

    for m, n in zip(mu, nu):
        if (m ** q + n ** q) > 1.0:
            n = (max(0.0, 1.0 - m ** q)) ** (1.0 / q)
        p = (max(0.0, 1.0 - m ** q - n ** q)) ** (1.0 / q)

        nu_adj.append(n)
        pi_adj.append(p)

    qrof_df = pd.DataFrame({
        f"qrof_mu_q{q}": mu,
        f"qrof_nu_q{q}": np.array(nu_adj),
        f"qrof_pi_q{q}": np.array(pi_adj)
    }, index=fuzzy_df.index)

    # Score and accuracy
    qrof_df[f"qrof_score_q{q}"] = (qrof_df[f"qrof_mu_q{q}"] ** q) - (qrof_df[f"qrof_nu_q{q}"] ** q)
    qrof_df[f"qrof_accuracy_q{q}"] = (qrof_df[f"qrof_mu_q{q}"] ** q) + (qrof_df[f"qrof_nu_q{q}"] ** q)

    return qrof_df

Q_RUNG = 3
qrof_train = qrof_feature_mapping(fuzzy_train, q=Q_RUNG)
qrof_val   = qrof_feature_mapping(fuzzy_val, q=Q_RUNG)
qrof_test  = qrof_feature_mapping(fuzzy_test, q=Q_RUNG)

print("\nq-ROF feature sample:")
display(qrof_train.head())

# ---------------------------
# 16. Combine deep-learning inputs + q-ROF fuzzy inputs
# ---------------------------
# For train data, because SMOTE changed samples, we first build
# combined tabular input using original X_train_scaled + qrof_train,
# then apply SMOTE jointly for consistency.

X_train_combined = np.hstack([X_train_scaled, qrof_train.values])
X_val_combined   = np.hstack([X_val_scaled, qrof_val.values])
X_test_combined  = np.hstack([X_test_scaled, qrof_test.values])

# Re-apply SMOTE on combined training representation
smote2 = SMOTE(random_state=SEED)
X_train_combined_sm, y_train_combined_sm = smote2.fit_resample(X_train_combined, y_train)

print("\nCombined train shape before SMOTE:", X_train_combined.shape)
print("Combined train shape after SMOTE :", X_train_combined_sm.shape)

# ============================================================
# OPTIONAL FOCAL LOSS
# ============================================================

# ---------------------------
# 17. Optional focal loss
# ---------------------------
USE_FOCAL_LOSS = True

def binary_focal_loss(gamma=2.0, alpha=0.25):
    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        epsilon = K.epsilon()
        y_pred = K.clip(y_pred, epsilon, 1.0 - epsilon)

        pt = tf.where(tf.equal(y_true, 1), y_pred, 1 - y_pred)
        alpha_t = tf.where(tf.equal(y_true, 1), alpha, 1 - alpha)

        return K.mean(-alpha_t * K.pow(1.0 - pt, gamma) * K.log(pt))
    return loss

loss_fn = binary_focal_loss(gamma=2.0, alpha=0.35) if USE_FOCAL_LOSS else "binary_crossentropy"

# ============================================================
# MODEL
# ============================================================

# ---------------------------
# 18. Build improved hybrid model
# ---------------------------
input_dim = X_train_combined_sm.shape[1]

inputs = layers.Input(shape=(input_dim,), name="Hybrid_Input")

x = layers.Dense(128, activation='relu')(inputs)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.30)(x)

x = layers.Dense(64, activation='relu')(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.25)(x)

x = layers.Dense(32, activation='relu', name="Deep_Fuzzy_Feature_Layer")(x)
x = layers.Dense(16, activation='relu')(x)

outputs = layers.Dense(1, activation='sigmoid', name="Stroke_Probability")(x)

model = models.Model(inputs=inputs, outputs=outputs, name="Hybrid_DeepLearning_qROFS_Model")

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss=loss_fn,
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc'),
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall')
    ]
)

model.summary()

# ---------------------------
# 19. Train model
# ---------------------------
early_stop = callbacks.EarlyStopping(
    monitor='val_auc',
    mode='max',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=4,
    min_lr=1e-6,
    verbose=1
)

history = model.fit(
    X_train_combined_sm, y_train_combined_sm,
    validation_data=(X_val_combined, y_val),
    epochs=80,
    batch_size=64,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

# ---------------------------
# 20. Plot learning curves
# ---------------------------
plt.figure(figsize=(7,4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Loss Curve")
plt.legend()
plt.show()

plt.figure(figsize=(7,4))
plt.plot(history.history['auc'], label='Train AUC')
plt.plot(history.history['val_auc'], label='Val AUC')
plt.title("AUC Curve")
plt.legend()
plt.show()

# ============================================================
# THRESHOLD TUNING
# ============================================================

# ---------------------------
# 21. Tune threshold on validation set
# ---------------------------
val_prob = model.predict(X_val_combined).ravel()

thresholds = np.arange(0.10, 0.91, 0.01)
best_threshold = 0.5
best_f1 = 0

threshold_results = []

for t in thresholds:
    val_pred = (val_prob >= t).astype(int)
    f1 = f1_score(y_val, val_pred, zero_division=0)
    prec = precision_score(y_val, val_pred, zero_division=0)
    rec = recall_score(y_val, val_pred, zero_division=0)

    threshold_results.append([t, prec, rec, f1])

    if f1 > best_f1:
        best_f1 = f1
        best_threshold = t

threshold_df = pd.DataFrame(threshold_results, columns=["Threshold", "Precision", "Recall", "F1"])
print("\nBest threshold:", best_threshold)
print("Best validation F1:", round(best_f1, 4))
display(threshold_df.sort_values("F1", ascending=False).head(10))

# ---------------------------
# 22. Test predictions
# ---------------------------
test_prob = model.predict(X_test_combined).ravel()
test_pred = (test_prob >= best_threshold).astype(int)

# ============================================================
# FINAL FUZZY DECISION LAYER
# ============================================================

# Here we combine:
# - feature-level qROF score
# - model probability
# for final fuzzy diagnosis

# ---------------------------
# 23. Final fuzzy decision fusion
# ---------------------------
test_qrof_score = qrof_test[f"qrof_score_q{Q_RUNG}"].values
test_qrof_mu    = qrof_test[f"qrof_mu_q{Q_RUNG}"].values
test_qrof_nu    = qrof_test[f"qrof_nu_q{Q_RUNG}"].values
test_qrof_pi    = qrof_test[f"qrof_pi_q{Q_RUNG}"].values

# Fuse deep probability + fuzzy feature score
# This makes the fuzzy system meaningful and not just a mirror of sigmoid output
final_fuzzy_score = (
    0.65 * test_prob +
    0.35 * ((test_qrof_score + 1) / 2.0)   # normalized from [-1,1] to [0,1]
)

# Tune fuzzy threshold on validation set too
val_qrof_score = qrof_val[f"qrof_score_q{Q_RUNG}"].values
val_final_fuzzy_score = (
    0.65 * val_prob +
    0.35 * ((val_qrof_score + 1) / 2.0)
)

best_fuzzy_threshold = 0.5
best_fuzzy_f1 = 0

for t in thresholds:
    val_fuzzy_pred = (val_final_fuzzy_score >= t).astype(int)
    f1 = f1_score(y_val, val_fuzzy_pred, zero_division=0)
    if f1 > best_fuzzy_f1:
        best_fuzzy_f1 = f1
        best_fuzzy_threshold = t

fuzzy_pred = (final_fuzzy_score >= best_fuzzy_threshold).astype(int)

print("\nBest fuzzy threshold:", best_fuzzy_threshold)
print("Best validation fuzzy F1:", round(best_fuzzy_f1, 4))

# ============================================================
# EVALUATION
# ============================================================

# ---------------------------
# 24. Deep learning metrics
# ---------------------------
print("\n===== Improved Hybrid Deep Learning Results =====")
print("Accuracy :", round(accuracy_score(y_test, test_pred), 4))
print("Precision:", round(precision_score(y_test, test_pred, zero_division=0), 4))
print("Recall   :", round(recall_score(y_test, test_pred, zero_division=0), 4))
print("F1-score :", round(f1_score(y_test, test_pred, zero_division=0), 4))
print("ROC-AUC  :", round(roc_auc_score(y_test, test_prob), 4))

print("\nClassification Report:")
print(classification_report(y_test, test_pred, zero_division=0))

print("Confusion Matrix:")
print(confusion_matrix(y_test, test_pred))

# ---------------------------
# 25. Final fuzzy metrics
# ---------------------------
print("\n===== Final q-ROFS Hybrid Diagnosis Results =====")
print("Accuracy :", round(accuracy_score(y_test, fuzzy_pred), 4))
print("Precision:", round(precision_score(y_test, fuzzy_pred, zero_division=0), 4))
print("Recall   :", round(recall_score(y_test, fuzzy_pred, zero_division=0), 4))
print("F1-score :", round(f1_score(y_test, fuzzy_pred, zero_division=0), 4))
print("ROC-AUC  :", round(roc_auc_score(y_test, final_fuzzy_score), 4))

print("\nClassification Report:")
print(classification_report(y_test, fuzzy_pred, zero_division=0))

print("Confusion Matrix:")
print(confusion_matrix(y_test, fuzzy_pred))

# ---------------------------
# 26. Save outputs
# ---------------------------
results = pd.DataFrame({
    "Actual": y_test.values,
    "DL_Probability": test_prob,
    "DL_Prediction": test_pred,
    f"qROF_mu_q{Q_RUNG}": test_qrof_mu,
    f"qROF_nu_q{Q_RUNG}": test_qrof_nu,
    f"qROF_pi_q{Q_RUNG}": test_qrof_pi,
    f"qROF_score_q{Q_RUNG}": test_qrof_score,
    "Final_Fuzzy_Score": final_fuzzy_score,
    "Fuzzy_Prediction": fuzzy_pred
})

SAVE_RESULTS = "/content/drive/MyDrive/Dataset /Fuzzy/stroke_hybrid_qrof_featurelevel_results.csv"
SAVE_MODEL   = "/content/drive/MyDrive/Dataset /Fuzzy/stroke_hybrid_qrof_featurelevel_model.keras"
SAVE_THRESH  = "/content/drive/MyDrive/Dataset /Fuzzy/stroke_threshold_tuning.csv"

results.to_csv(SAVE_RESULTS, index=False)
threshold_df.to_csv(SAVE_THRESH, index=False)
model.save(SAVE_MODEL)

print(f"\nSaved results to: {SAVE_RESULTS}")
print(f"Saved threshold tuning to: {SAVE_THRESH}")
print(f"Saved model to: {SAVE_MODEL}")

display(results.head(10))
```
